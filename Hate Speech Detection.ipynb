```python
import pandas as pd

# Add example data
data = [
    [1, 1, 0, 0, 0, "I hate you"],
    [2, 0, 1, 0, 1, "You are so stupid"],
    [3, 0, 0, 1, 2, "Have a nice day!"],
    [4, 0, 1, 0, 1, "You're an idiot"],
    [5, 1, 0, 0, 0, "I wish you didn't exist"],
    [6, 0, 0, 1, 2, "Good morning, everyone"]
]

columns = ["count", "hate_speech", "offensive_language", "neither", "class", "Tweet"]

df = pd.DataFrame(data, columns=columns)
df.to_csv("twitter.csv", index=False)

print("twitter.csv filled with sample data.")
```

```python
df = pd.read_csv("twitter.csv")
df.head()
```

```python
from sklearn.model_selection import train_test_split

X = df['Tweet']
y = df['class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(X_train_vec, y_train)
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test_vec)
print(classification_report(y_test, y_pred, zero_division=0))  # avoids divide-by-zero warning
```

```python
custom_text = ["You are dumb", "Hope you're doing well"]
custom_vec = vectorizer.transform(custom_text)
print(model.predict(custom_vec))
```
